```python
# Import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import SubplotSpec
from matplotlib import style
import seaborn as sns
```

```python
# Load the data from a csv file
perf_metrics_mod_allreg = pd.read_csv('Results/Results_best_classifiers_modeling_per_region_all.csv', index_col=0)
perf_metrics_modft2_allreg = pd.read_csv('Results/Results_best_classifiers_modeling_ft2_per_region_all.csv', index_col=0)
perf_metrics_fmri_allreg = pd.read_csv('Results/Results_best_classifiers_fmri_per_region_all.csv', index_col=0)
```

```python
# Select values from the first row
perf_metrics_mod_allreg
```

```python
perf_metrics_modft2_allreg
```

```python
perf_metrics_fmri_allreg
```

```python
#set tick labels font to size 11
plt.rc('xtick', labelsize=11) 
plt.rc('ytick', labelsize=11) 

# Define the figure size
plt.figure(figsize=(10, 6))

# Using the style for the plot
sns.set_style("ticks")

# Create a line plot for the accuracy
plt.plot(perf_metrics_mod_allreg.iloc[0], marker='o', label='Modeling dataset', linewidth=2, color='lightcoral')
plt.plot(perf_metrics_modft2_allreg.iloc[0], marker='x', label='Modeling dataset + FT2', linewidth=2, color='green')
plt.plot(perf_metrics_fmri_allreg.iloc[0], marker='^', label='fMRI dataset', linewidth=2, color='steelblue')
plt.legend(fontsize=12.5, loc='lower right')
plt.xlabel('Auditory Regions', fontsize=12.5)
plt.ylabel('Accuracy', fontsize=12.5)
#plt.title('Accuracy for all regions')

# Adjust plot margins
plt.tight_layout()

# Save the plot
plt.savefig('Results/Lineplot_acc_allmodels_and_regions.png', bbox_inches='tight')
```
